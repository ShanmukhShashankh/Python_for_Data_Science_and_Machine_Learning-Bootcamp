```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(-10,10,100)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph')
```

```python
y=np.cos(x)
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y+50,'b.')
plt.subplot(2,2,3)
plt.plot(x+50,y)
plt.subplot(2,2,4)
plt.plot(x+50,y+50)
```

```python
fig = plt.figure(figsize=(10,8))

gph1 = fig.add_axes([0,0,2,1])
gph2 = fig.add_axes([0,-1.1,2,1])

gph1.plot(x,y)
gph1.set_xlabel('X')
gph1.set_ylabel('Xi')

gph2.plot(x,y**2)
gph2.set_xlabel('X')
gph2.set_ylabel('|Xi|^2')

plt.tight_layout()
```

```python
tim = plt.figure()

mit = tim.add_axes([0,0,4,2])

mit.plot(x,y)
mit.plot(x,y**2)
```
